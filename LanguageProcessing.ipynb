# Pronoun Resolution

Alway there will be a ambiguity what they are referring at.

```
The thieves stole the paintings. They were subsequently found.
```

Here what they referring at -- Thieves or painting.  

To tackle such problem we have 
1. "Anaphora resolution. -->  what the pronoun or noun referring at."
2. "Semantic Resolution" --> how a noun relates to the verb
