# Chunking



here we combine the some POS into phrase, this will help to entity recognition.

```python
import nltk
```

## Noun Phrase Chunking

```python
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),
 ("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
```

```python
grammar = "NP: {<DT>?<JJ>*<NN>}"
#? Noun phrase start with Determiner , it can have multiple adjective and end with Noun
```

```python
chunk_parser = nltk.RegexpParser(grammar)
result = chunk_parser.parse(sentence)
```

```python
print(result)
```

```python
result.draw()
```
