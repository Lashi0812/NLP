# Perceptron

```python
import torch
from torch import nn
from torch.nn import functional as F 


import matplotlib.pyplot as plt
```

```python
class Perceptron(nn.Module):
    def __init__(self,input_dim) -> None:
        super().__init__()
        self.fc1 = nn.Linear(in_features=input_dim,out_features=1)
    
    def forward(self,x):
        return torch.sigmoid(self.fc1(x)).squeeze()
```

# Activation function

## Sigmoid

```python
x = torch.arange(-5,5,0.1)
y = torch.sigmoid(x)
plt.plot(x,y);
```

## tanh

```python
x = torch.arange(-5,5,0.1)
y = torch.tanh(x)
plt.plot(x,y);
```

## relu

```python
x = torch.arange(-5,5,0.1)
y = torch.relu(x)
plt.plot(x,y);
```

## Leaky relu

```python
x = torch.arange(-5,5,0.1)
y = F.leaky_relu(x,negative_slope=0.1)
plt.plot(x,y);
```

## prelu

```python
x =torch.arange(-5,5,0.1)
m = nn.PReLU()
y = m(x)
plt.plot(x,y.detach().numpy());
```
